## Detecting Overfitting and Underfitting using learning curves

### We'll be using 3 Models for this exercise
>* Decision Tree Model
>* Logistic Regression Model
>* Support Vector Machine Model

### One of the model underfits other overfits and one fits just right we'll find each of them using __learning curves__

```python
#import read and split data
import pandas as pd
import numpy as np 

data = pd.read_csv("data2-3.csv")
X = np.array(data[['x1','x2']])
y = np.array(data['y'])
```

```python
#fix random seed
np.random.seed(55)
```

```python
#importing models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
```

### Random function and Plotting function

```python
#random function
def randomize(X,Y):
    permutation = np.random.permutation(Y.shape[0])
    X2 = X[permutation,:]
    Y2 = Y[permutation]
    return X2, Y2
```

For the function below:
* **estimator** is the classifier which tells the used model
* **X and y** is our data split into features and labels
* **train_sizes**: size of chunks of data used to draw each point
* **train_scores**: training scores for each chunk of data
* **test_scores**: test scores for each chunk of data

```python
#plotting function
import matplotlib.pyplot as plt 
from sklearn.model_selection import learning_curve

def draw_learning_curves(X, y, estimator, num_trainings):
    train_sizes, train_scores, test_scores = learning_curve(estimator, X2, y2, cv=3, n_jobs=1, train_sizes= np.linspace(0.1,1,num_trainings)) 
    
    train_scores_mean = np.mean(train_scores, axis=1)   #mean 
    train_scores_std = np.std(train_scores, axis=1)     #standard deviation
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    
    plt.grid()
    
    plt.title("Learning Curves")
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    
    plt.plot(train_scores_mean, 'o-', color="g",
             label="Training score")
    plt.plot(test_scores_mean, 'o-', color="y",
             label="Cross-validation score")


    plt.legend(loc="best")

    plt.show()
```

```python
# Decision Tree
estimator = GradientBoostingClassifier()

X2,y2 = randomize(X,y)
estimator.fit(X2,y2)
draw_learning_curves(X2,y2,estimator,20)
```
